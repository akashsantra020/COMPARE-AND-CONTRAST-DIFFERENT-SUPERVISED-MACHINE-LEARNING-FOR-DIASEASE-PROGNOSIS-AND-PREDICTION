```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr(method='spearman')
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
# Define the function to be optimized (accuracy of decision tree classifier)
def dt_score(max_depth, max_features):
    # Create a decision tree classifier with the specified hyperparameters
    dt = DecisionTreeClassifier(max_depth=int(max_depth), max_features=min(max_features, 0.999))
    
    # Fit the classifier to the training data
    dt.fit(X_train, y_train)
    
    # Predict the classes of the testing data
    y_pred = dt.predict(X_test)
    
    # Calculate the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Return the accuracy of the classifier (sum of diagonal elements of confusion matrix)
    return np.sum(np.diag(cm)) / np.sum(cm)
```

```python
# Define the search space for the hyperparameters
params = {'max_depth': (2, 10),
          'max_features': (0.1, 0.999)}
```

```python
# Create the Bayesian optimization object and run the optimization
dt_bo = BayesianOptimization(dt_score, params)
dt_bo.maximize(n_iter=10, init_points=5)
```

```python
# Print the best hyperparameters and corresponding accuracy
print('Best hyperparameters: ', dt_bo.max['params'])
print('Best accuracy: ', dt_bo.max['target'])

# Create a decision tree classifier with the best hyperparameters
best_dt = DecisionTreeClassifier(max_depth=int(dt_bo.max['params']['max_depth']), 
                                 max_features=min(dt_bo.max['params']['max_features'], 0.999))
```

```python
# Define the objective function to optimize
def objective(params):
    # Initialize the decision tree classifier with the hyperparameters
    clf = DecisionTreeClassifier(max_depth=int(params[0]), min_samples_split=int(params[1]), 
                                  min_samples_leaf=int(params[2]), random_state=0)
    
    # Train the decision tree classifier on the training data
    clf.fit(X_train, y_train)
    
    # Predict the labels of the test data using the trained classifier
    y_pred = clf.predict(X_test)
    
    # Compute the confusion matrix and return it as the objective value
    cm = confusion_matrix(y_test, y_pred)
    return np.sum(-cm.diagonal())  # maximize the sum of correctly classified samples
```

##Bayesian Optimization for SVM(Linear)

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from skopt import gp_minimize
from skopt.space import Real, Integer
from skopt.utils import use_named_args
```

```python
# Define the search space for hyperparameters
space = [Real(0.001, 100.0, name='C'),
         Real(0.001, 10.0, name='gamma')]
```

```python
# Define the objective function to optimize
@use_named_args(space)
def objective(**params):
    # Initialize the SVM classifier with the hyperparameters
    clf = SVC(kernel='linear', **params)
    
    # Train the SVM classifier on the training data
    clf.fit(X_train, y_train)
    
    # Predict the labels of the test data using the trained classifier
    y_pred = clf.predict(X_test)
    
    # Compute the confusion matrix and return it as the objective value
    cm = confusion_matrix(y_test, y_pred)
    return -np.sum(cm.diagonal())  # minimize the sum of misclassified samples
```

```python
# Perform Bayesian optimization using the GP-based optimizer
res = gp_minimize(objective, space, n_calls=20, random_state=0)
```

```python
# Print the optimal hyperparameters and objective value
print("Optimal Hyperparameters: ", res.x)
print("Optimal Objective Value: ", -res.fun)
```
